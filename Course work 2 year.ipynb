```python
def Starter(G): #Функция создания начального графа
    G.add_edge(1,2)
    G.add_edge(2,3)
    G.add_edge(3,4)
    G.add_edge(4,5)
    G.add_edge(4,2)
```

```python
def Add_Edge(G,s,e): #Добавить грань
    plt.clf()
    G.add_edge(s,e)
    nx.draw_networkx(G, with_labels = True) 
    
```

```python
def Adjacency_matrix(G): # Матрица смежности
    A=nx.adjacency_matrix(G)
    print(A.todense() )
```

```python
def Incidence_matrix(G): #Матрица инцидентности
    I=nx.incidence_matrix(G)
    print(I.todense())
```

```python
def searchWidth(G): #Обход в ширину
    bfs_tree = nx.bfs_tree(G, source=1)
    print("Bfs tree nodes:", list(bfs_tree.nodes()))
```

```python
def searchDepth(G): #Обход в глубину
    dfs_tree = nx.dfs_tree(G, source=1)
    print("DFS tree nodes:", list(dfs_tree.nodes()))
```

```python
def ShortestPath(G,source=0,target=0):#Поиск кратчайшего пути
    s_p=nx.shortest_path(G, source, target)
    print(s_p)
```
